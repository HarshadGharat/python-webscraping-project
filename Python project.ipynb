```python
page=requests.get("https://csmia.adaniairports.com/flights.aspx")
page
```

```python
soup=BeautifulSoup(page.content)
print(soup.prettify())
```

```python

Arrival=soup.find_all("div",class_="sticky-head-table")
Arrival_Flights=[]
for i in Arrival:
    Arrival_Flights.append(i.text)
Arrival_Flights
```

```python
serial=soup.find_all("td",class_="pdsc")
serialno=[]
for i in serial:
    serialno.append(i.text)
serialno
```

```python
srno=serialno[0:len(serialno):10]
srno
```

```python
flight=serialno[2:len(serialno):10]
flight
```

```python
Destination=serialno[3:len(serialno):10]
Destination
```

```python
Date=serialno[4:len(serialno):10]
Date
```

```python
SCH=serialno[5:len(serialno):10]
SCH
```

```python
Calling_time=serialno[6:len(serialno):10]
Calling_time
```

```python
Status=serialno[7:len(serialno):10]
Status
```

```python
Terminal=serialno[8:len(serialno):10]
Terminal
```

```python
import pandas as pd
```

```python
Domasticflight=pd.DataFrame({})
Domasticflight["SR.NO."]=srno
Domasticflight["FLIGHT"]=flight
Domasticflight["DESTINATIN"]=Destination
Domasticflight["DATE"]=Date
Domasticflight["SCH"]=SCH
Domasticflight["CALLING TIME"]=Calling_time
Domasticflight["STATUS"]=Status
Domasticflight["TERMINAL"]=Terminal

Domasticflight
```

```python
Domasticflight.to_csv("Domasticflights")
```
